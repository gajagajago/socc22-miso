```python
import json
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('miso/active_jobs_per_gpu.json') as f:
    data_miso = json.load(f)
with open('full/active_jobs_per_gpu.json') as f:
    data_full = json.load(f)    
with open('static/active_jobs_per_gpu.json') as f:
    data_static = json.load(f)        
# with open('oracle/active_jobs_per_gpu.json') as f:
#     data_oracle = json.load(f)            
```

```python
plt.plot(data_miso, label='miso')
plt.plot(data_full, label='full')
plt.plot(data_static, label='static')
# plt.plot(data_oracle, label='oracle')
plt.legend()
plt.grid(axis='both', ls='dotted')
```

```python
with open('miso/progress.json') as f:
    data_miso = json.load(f)
with open('full/progress.json') as f:
    data_full = json.load(f)    
with open('static/progress.json') as f:
    data_static = json.load(f)     
```

```python
len(data_miso), len(data_full), len(data_static)
```

```python
plt.plot(data_miso, label='miso')
plt.plot(data_full, label='full')
plt.plot(data_static, label='static')
# plt.plot(data_oracle, label='oracle')
plt.legend()
plt.grid(axis='both', ls='dotted')
```

```python
step = 1
dmiso = [data_miso[i]-data_miso[i-step] for i in range(step,len(data_miso),step)]
dfull = [data_full[i]-data_full[i-step] for i in range(step,len(data_full),step)]
dstatic = [data_static[i]-data_static[i-step] for i in range(step,len(data_static),step)]
```

```python
plt.plot(dmiso, label='miso')
plt.plot(dfull, label='full')
plt.plot(dstatic, label='static')
# plt.plot(data_oracle, label='oracle')
plt.legend()
plt.grid(axis='both', ls='dotted')
```
